```python
import numpy as ny
import math

mean = 0
test_scores2 = ny.array([5, 6, 7, 8, 9])

mean = sum(test_scores2) / len(test_scores2)
mean = int(mean)

squared_difference = []

for x in test_scores2:
    sd = math.pow(x - mean, 2)   
    squared_difference.append(sd)  

print(mean)
print(squared_difference)

variance = sum(squared_difference) / len(test_scores2)

standard_deviation2 = math.pow(variance, 0.5)

print("Variance:", variance)
print("Standard Deviation:", standard_deviation2)
```
